```python
import numpy as np
import matplotlib.pyplot as plt

# pcm_1 = '../data/a.pcm'  # sv56demo input
pcm_1 = '../output/saturation.pcm'  # sv56demo input
pcm_2 = '../sv56demo_output/saturation_out.pcm'  # sv56demo output 
sr = 8000
dtype = np.int16 

# PCM 파일 읽기

def read_pcm(pcm_file):
    with open(pcm_file, "rb") as pcm_in:
        return np.frombuffer(pcm_in.read(), dtype=dtype)


pcm_data1 = read_pcm(pcm_1)
pcm_data2 = read_pcm(pcm_2)


# 샘플 인덱스 생성
samples = np.arange(len(pcm_data1)) / sr  # 초 단위로 변환

# 파형 비교 그래프
plt.figure(figsize=(12, 5))
plt.plot(samples, pcm_data1, label="Original", alpha=0.7)
plt.plot(samples, pcm_data2, label="SV56 Processed", alpha=0.7, linestyle="--")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.title("Saturation Comparison")
plt.legend()
plt.grid()
plt.show()
```
